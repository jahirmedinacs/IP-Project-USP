# Generating images


```python
def sound_2_image(sound_data, matrix_shape, matrix_patter, rang=[-(1<<15), (1<<15) -1]):
    image_carry = np.zeros(matrix_shape, np.uint8)
    for index_bean in range(len(sound_data)):
        [x_i, y_i] = matrix_patter[index_bean]

        wavelength = mlb.Int2nM(sound_data[index_bean], rang)

        image_carry[x_i, y_i, :] =  mlb.wavelength_to_rgb(wavelength)

#         image_carry = image_carry.astype(np.uint8)

    return image_carry
```

```python
dinamic = False

hilbert_shape = (1<<12, 1<<12, 3)
hilbert_path = "../hilbert_curve/pattern_12_size.npy"
hilbert_pattern = np.load(hilbert_path)

for item_id in range(len(files_wav)):
   
    print(">>\tSTATE\t: {:d}".format(item_id + 1))
    pprint.pprint(files_wav[item_id])
    print("")
    print(meta_data.iloc[[item_id]])
        
    [carry_l, carry_r] = mlb.stereo_chanel_data(sound_data_container[item_id][0])
    
    if dinamic:
        carry_len = np.array(meta_data.iloc[[item_id]]).item(-3) - 2
        near_2_pow = int(np.ceil(np.log2(dummy_len)) // 2)
        size_len = 1<<near_2_pow
        hilbert_shape = (size_len, size_len, 3)
        hilbert_path = "../hilbert_curve/pattern_{:d}_size.npy".format(near_2_pow)
    else:
        pass
        
    print("Hilberth Pattern\t", hilbert_shape)

    ### L
    matrx = sound_2_image(carry_l, hilbert_shape, hilbert_pattern)
    print(matrx)
    scipy.misc.toimage(matrx, cmin=0, cmax=255).save(DATA_PATH + files_wav[item_id][:-4] + "_{:s}.png".format('L'))
    
    ### R
    scipy.misc.toimage(
        sound_2_image(carry_r, hilbert_shape, hilbert_pattern),
        cmin=0, cmax=255).save(DATA_PATH + files_wav[item_id][:-4] + "_{:s}.png".format('R'))
```
