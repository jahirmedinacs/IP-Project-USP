```python
def data_2_wav(data, specs, path, verbose=False):
    os.system("touch " + path)
    if verbose:
        print("creating destination file\t" + path)
    
    try:           
        output.close()
    except:
        if verbose:
            print("alocating file")
            
        output = wave.open(path, 'w')
    else:
        output = wave.open(path, 'w')
        if verbose:
            print("file already exists")
    
    if verbose:
            print("setting sound specs")
    
    specs[0] = 1
    specs[3] = len(data)
    
    output.setparams(specs)
    
    if verbose:
        print("formating data...")
        
    _carry = []
    for ii in list(data):
        _carry.append(struct.pack('<h', ii))

    _carry = b''.join(_carry)
    
    if verbose:
            print("writing...")
    
    output.writeframes(_carry)
    
    output.close()
    
    if verbose:
        print("DONE")
```

```python
DATA_PATH = "../data/refined-data/"

general_specs = meta_data.values[0][2:]
general_specs[0] = 1
general_specs[3] = len(data_l)

SECS_PER_SAMPLE = 5

execed = general_specs[3] % 44100
sec_temp = ( general_specs[3] - execed ) // 44100
sec_temp_execed = sec_temp % SECS_PER_SAMPLE
samples_amount =  ( sec_temp - sec_temp_execed ) // SECS_PER_SAMPLE

print(samples_amount)

frames_execed = sec_temp_execed * 44100

print(frames_execed, sec_temp_execed)
```

```python
import noice_inductor

frames_per_sec = general_specs[2]
NAME = "tank_sample_"

data_r = list(data_r)
data_l = list(data_l)

print("RIGTH CHANNEL SAMPLES , RANDOM NOISE DIFFRENT %")
noices_samples_r = []
noices_samples_r.append([noice_inductor.random_noise(data_r, 10), "random_10"])
noices_samples_r.append([noice_inductor.random_noise(data_r, 20), "random_20"])
noices_samples_r.append([noice_inductor.random_noise(data_r, 30), "random_30"])
noices_samples_r.append([noice_inductor.random_noise(data_r, 40), "random_35"])

print("LEFT CHANNEL SAMPLES , RANDOM NOISE DIFFRENT %")
noices_samples_l = []
noices_samples_l.append([noice_inductor.random_noise(data_l, 10), "random_10"])
noices_samples_l.append([noice_inductor.random_noise(data_l, 20), "random_20"])
noices_samples_l.append([noice_inductor.random_noise(data_l, 30), "random_30"])
noices_samples_l.append([noice_inductor.random_noise(data_l, 40), "random_35"])
```

```python

ref_start = 0
for i in range(1, samples_amount + 1, 1):
    ref_end = frames_per_sec * i * SECS_PER_SAMPLE
    
    """ RIGHT """
    
    r_data = data_r[ ref_start : ref_end ]
    
    data_2_wav(r_data, general_specs,
               DATA_PATH + NAME +
               "pure_" + str(i) + "_R.wav", True )
    
    for sub_sample in noices_samples_r:
        data_2_wav(sub_sample[0][ ref_start : ref_end ], general_specs,
               DATA_PATH + NAME +
               sub_sample[1] + "_" + str(i) + "_R.wav", True )
    
    """ LEFT """
    
    l_data = data_l[ ref_start : ref_end ]
    
    data_2_wav(l_data, general_specs,
               DATA_PATH + NAME +
               "pure_" + str(i) + "_L.wav", True )
    
    for sub_sample in noices_samples_l:
        data_2_wav(sub_sample[0][ ref_start : ref_end ], general_specs,
               DATA_PATH + NAME +
               sub_sample[1] + "_" + str(i) + "_L.wav", True )
    
    
    print( (ref_end - ref_start) / frames_per_sec , " DONE STEP :\t", i )
    
    ref_start = frames_per_sec * i * SECS_PER_SAMPLE
    
```
