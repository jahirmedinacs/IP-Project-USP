```python
DATA_PATH = "../data/"
files_png = [ file for file in os.listdir(DATA_PATH) if file[-4:] == ".png"]


image_data_container = {}
for png_f in files_png:
    try:
        image_data_container[int(png_f[:2])] = [imr(DATA_PATH + png_f),png_f]
    except:
        print("ERROR FILE {:s}".format(png_f))
```

```python
pprint.pprint(image_data_container)
```

```python
def kernel(image_data, id_container, data_container):
    print("KERNEL {:d} STARTED".format(id_container))
    image_data_container[id_container].append(des.descriptors(image_data))
    print("KERNEL DONE")
```

```python
thread_container = [None for _ in range(len(image_data_container))]

for index in range(len(image_data_container)):
    thread_container[index] = threading.Thread(target=kernel,args=(
        image_data_container[index + 1][0],
        index + 1,
        image_data_container))

for thrd in thread_container:
    thrd.start()
```
