```python
import os
import sys

def add_to_path(str):
    try:
        sys.path.index(str)
    except:
        print(str + "\t is already in the path list")
    else:
        sys.path.append(str)
        print(str + "\t add to the path list")

# this add all the libraris and scripts made for this project

add_to_path('../datalib/')
add_to_path('../khinsider-master/')
```
